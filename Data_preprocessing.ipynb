```python
# 在原始表格中添加新列 Is_Host
medal_table['Is_Host'] = np.nan

medal_table['Is_Host'] = medal_table.apply(
    lambda row: row['NOC'] == host_data.get(row['Year'], ''), axis=1
)

medal_table_year = {}
for i in range(medal_table.shape[0]):
    row = medal_table.iloc[i]  # 获取第i行数据
    year = row['Year']  # 获取当前行的年份

    # 如果该年份还未在字典中，初始化一个空列表
    if year not in medal_table_year:
        medal_table_year[year] = []

    # 将当前行数据添加到对应年份的列表中
    medal_table_year[year].append(row)

# for year, data in medal_table_year.items():
# # 打印结果
for year, data in medal_table_year.items():
    print(f"Year: {year}")
    print(pd.DataFrame(data))  # 将列表转换为DataFrame以便更好地显示
    print("\n")
```
